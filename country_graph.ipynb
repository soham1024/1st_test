```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_data=pd.read_csv('countries.csv')
```

```python
number=0
initial=-1
while(number<1704):
    ending=number
    if(ending<1692):
        ending+=12
    else:
        ending=1703
    number+=12
    # x axis values 
    x = file_data['year'][initial+1:ending]
    # corresponding y axis values 
    y = file_data['population'][initial+1:ending]
  
    # plotting the points  
    plt.plot(x, y)
  
    # naming the x axis 
    plt.xlabel('years') 
    # naming the y axis 
    plt.ylabel('populaton') 
  
    # giving a title to my graph 
    plt.title(file_data['country'][ending])
  
    # function to show the plot 
    plt.show()
    initial=ending
```
