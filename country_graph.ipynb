```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_data=pd.read_csv('countries.csv')
```

```python
n=0
m=-1
while(n<1704):
    i=n
    if(i<1692):
        while(file_data['country'][i]==file_data['country'][i+1]):
            i+=1
    else:
        i=1703
    n+=12
    # x axis values 
    x = file_data['year'][m+1:i]
    # corresponding y axis values 
    y = file_data['population'][m+1:i]
  
    # plotting the points  
    plt.plot(x, y)
  
    # naming the x axis 
    plt.xlabel('years') 
    # naming the y axis 
    plt.ylabel('populaton') 
  
    # giving a title to my graph 
    plt.title(file_data['country'][i])
  
    # function to show the plot 
    plt.show()
    m=i
```
